```python
import pandas as pd 
from sqlalchemy import create_engine

df = pd.read_csv('sample_sales.csv')
engine = create_engine('sqlite:///sample_sales.db', echo=False)
df.to_sql(name='sample_sales', con= engine)
sql = "select * from sample_sales limit 10"
df_sample = pd.read_sql(sql, engine)
```
